```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
data.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data.year, data.capital_income,color='red',marker='+')
```

```python
year_c = data.drop('capital_income',axis='columns')
income_c = data.capital_income
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(year_c,income_c)
```

```python
print('Answer : ' , reg.predict([[2020]]))
reg.predict([[2016]])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
# plt.scatter(data.year, data.capital_income,color='red',marker='+')
plt.plot(data.year, data.capital_income,color='red',marker='+')
plt.plot(data.year,reg.predict(data[['year']]),color='blue')
```
